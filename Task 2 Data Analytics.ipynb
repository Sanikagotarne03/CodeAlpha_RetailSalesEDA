```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.boxplot(data=df[['Sales', 'Profit', 'Discount']])
plt.title("Outliers in Sales, Profit, and Discount")
plt.show()
```

5. Outlier Observation
Sales and Profit have clear outliers, meaning some orders have very high or very low values.  
Discount has fewer outliers and is mostly normal.

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Category-wise Total Sales')
plt.ylabel('Total Sales')
plt.xlabel('Category')
plt.show()
```

6. Category-wise Sales Observation
Technology category has the highest total sales, followed by Furniture and Office Supplies.

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Profit vs Sales')
plt.show()
```

7.Profit vs Sales Observation
There is a weak positive relationship between sales and profit. Some orders have high sales but low or even negative profit.

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.show()
```

8.Discount vs Profit Observation
Higher discounts generally reduce profit. Many discounted orders show low or even negative profit.

```python
df['Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Customer Segment Distribution')
plt.ylabel('')
plt.show()
```

9.Customer Segment Observation
The Consumer segment has the highest number of customers, followed by Home Office. Corporate segment is the smallest.

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_df = df[['Sales', 'Profit', 'Discount']]  # Only numeric columns
plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

10.Correlation Heatmap Observation
Sales and Profit have a weak positive relationship.  
Discount has a weak negative relationship with Profit.  
Sales and Discount have almost no relation.

Final Summary and Insights

From the analysis, we observed the following:

- The Technology category has the highest sales.
- The Central region shows the highest profit.
- The Consumer segment makes the maximum purchases.
- Higher discounts generally reduce profit.
- High sales do not always mean high profit because profit also depends on discount and product type.

Other observations:
- There is a weak positive relationship between sales and profit.
- Some outliers exist in sales and profit, which indicate unusually high or low values.
- There are no missing or duplicate values in the dataset.

Conclusion:
To improve profit, the business should focus more on Technology products, Consumer customers, and avoid giving very high discounts.
